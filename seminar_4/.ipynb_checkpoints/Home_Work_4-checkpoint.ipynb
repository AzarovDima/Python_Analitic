###### Построим коробочки с усами от кол-ва Ванн и цены домов

```python
plt.figure(figsize=(16, 8))

sns.boxplot(x=df['bathrooms'], y=df['price'], whis=1.5)

plt.xlabel('Ванны')
plt.ylabel('Цена')
plt.title('Зависимость стоимости домов от кол-ва Ванн');
```

Выводы:
Стоимость домов увеличивается по мере увеличения кол-ва Ванн в домах

```python
plt.figure(figsize=(6, 4))

sns.scatterplot(x=df['grade'], y=df['price'])

plt.title("График изменения цены от оценки домов")
plt.xlabel('Оценка')
plt.ylabel('Цена');
```

Вывод: Чем выше оценка тем дороже может стоить дом.

```python
sns.jointplot(x=df['lat'], y=df['long'], kind='kde');
```

```python
sns.jointplot(x=df['lat'], y=df['price'], kind='reg');
```

```python
sns.jointplot(x=df['long'], y=df['price'], kind='reg');
```

Вывод:
1. Чем Севернее в городе df['lat'] > 47.5 тем дома дороже
2. Чем Западнее в городе df['long'] < 121.8 тем дома дороже

```python
plt.figure(figsize=(16, 8))

sns.boxplot(x=df['view'], y=df['price'], whis=1.5)

plt.xlabel('Вид дома(или из окна))))')
plt.ylabel('Цена')
plt.title('Зависимость стоимости домов от вида дома');
```

Вывод:
Чем лучше вид из окна или вид дома самого))), тем дороже стоит дом

```python
sns.jointplot(x=df['grade'], y=df['price'], kind='reg');
```

Вывод!
Чем лучше конструкция и дизайн тем дома стоят дороже!
